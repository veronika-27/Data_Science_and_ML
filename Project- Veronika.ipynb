```python
pet_data=pd.read_csv('CensusProfile2016.csv',encoding='cp863', skiprows=[0,2])
```

```python
pet_data.head()
```

```python
pet_data.set_index('Topic', inplace=True)
```

```python
pet_new=pet_data.iloc[1896:1919, :]
pet_new.head()
```

```python
pet_new.reset_index(inplace=True)
```

```python
pet_new.columns
```

```python
pet_nn=pet_new.drop(columns=['Topic','Note', 'Male', 'Female', 'Total.1', 'Male.1', 'Female.1'])
```

```python
pet_nn.head()
```

```python
pet_n=pet_nn.iloc[3:, :]
pet_n.head()
```

```python
pet_n['Total'].sum()
```

```python
type(pet_n['Total'].sum())
```

```python
pet_n['Proportions (%)']= pet_n['Total']/pet_n['Total'].sum()*100
```

```python
pet_n.rename(columns={'Characteristics':'Industry', 'Total':'Workforce'}, inplace=True)
```

```python
pet_n.sort_values(by='Proportions (%)', ascending=False)
```

```python
# Importing re package for using regular expressions
import re
  
# Function to clean the names
def Clean_ind(Industry_name):
    # Search for digit in the ind. name followed by space
    
    if re.search('[0-9]\s', Industry_name):
  
        # Extract the position of the end of pattern
        pos = re.search('[0-9]\s', Industry_name).end()
  
        # return the cleaned name
        return Industry_name[pos:]
  
    else:
        # if clean up needed return the same name
        return Industry_name
          
# Updated the industry columns
pet_n['Industry'] = pet_n['Industry'].apply(Clean_ind)
  
# Print the updated dataframe
pet_n
```

```python
pet_n['Industry']=pet_n['Industry'].str.title()
```

```python
pet_n['Industry']=pet_n['Industry'].str.replace(';' , ',')
```

```python
pet_n['Industry']=pet_n['Industry'].str.replace('And' , '&')
```

```python
pet_n['Industry']=pet_n['Industry'].str.replace('(', ',')
pet_n['Industry']=pet_n['Industry'].str.replace(')' , '')
```

```python
pet_n['Industry'].values
```

```python
ann_data['Industry'].count()
```

```python
pet_n.replace({"Other Services ,Except Public Administration" : "Other Services, Except Public Administration",
              'Administrative & Support, Waste Management & Remediation Services':'Administrative & Support & Waste Management Services', 
               'Arts, Entertainment & Recreation':'Arts, Entertainment, & Recreation', 'Administrative & Support, Waste Management & Remediation Services':'Administrative & Support & Waste Management Services',
              'Professional, Scientific & Technical Services':'Professional, Scientific, & Technical Services', 'Information & Cultural Industries': 'Information', 
              'Management Of Companies & Enterprises': 'Management of Companies & Enterprises'}, inplace=True)
```

```python
df_comp = pd.merge(ann_data, pet_n, on='Industry', how='inner')
df_comp.rename(columns={'Proportions (%)_x':'Ann_Arbor', 'Proportions (%)_y': 'Peterborough'}, inplace=True)
```

```python
df_comp.drop(columns={'Workforce_x',  'Workforce_y'}, inplace=True)
```

```python
df_comp=df_comp.sort_values(['Ann_Arbor', 'Peterborough'], ascending=False)
df_comp.set_index('Industry')
```

```python
import matplotlib.pyplot as plt

x = df_comp['Industry'].values
Ann_Arrbor = df_comp['Ann_Arbor'].values
Peterborough=df_comp['Peterborough'].values

fig, axs = plt.subplots(1,2, figsize=(10,4), sharey=True, sharex=True)
axs[0].barh(x,Ann_Arrbor,  label='Ann_Arbor')
axs[1].barh(x, Peterborough,color='darkorange', label= 'Peterborough')
axs[0].legend()
axs[1].legend()
ax=df_comp.plot.bar(x='Industry',figsize=(12,4))
plt.ylabel('Share of industries (%)')

fig.suptitle('Employment by Industries (2016)', alpha=0.8)
```
